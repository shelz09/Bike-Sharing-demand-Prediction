## 3. Data Preparation

```python
# # Get the dummy variables for month, season, weathersit, weekday and Let's drop the first column from  using 'drop_first = True'pd.get_dummies(bike.season,drop_first=True)
months=pd.get_dummies(ds.Month,drop_first=True)
weekdays=pd.get_dummies(ds.weekday,drop_first=True)
weather_sit=pd.get_dummies(ds.weathersit,drop_first=True)
seasons=pd.get_dummies(ds.season,drop_first=True)

#bike=pd.concat([seasons,bike], axis=1)
```

```python
# Add the results to the original bike dataframe
ds=pd.concat([months,weekdays,weather_sit,seasons,ds],axis=1)
ds.head()
```

```python
#changing the data types
#months
ds['Aug'] = ds['Aug'].astype(int)
ds['Dec'] = ds['Dec'].astype(int)
ds['Feb'] = ds['Feb'].astype(int)
ds['Jan'] = ds['Jan'].astype(int)
ds['July'] = ds['July'].astype(int)
ds['June'] = ds['June'].astype(int)
ds['Mar'] = ds['Mar'].astype(int)
ds['May'] = ds['May'].astype(int)
ds['Nov'] = ds['Nov'].astype(int)
ds['Oct'] = ds['Oct'].astype(int)
ds['Sep'] = ds['Sep'].astype(int)
#weekdays
ds['Mon'] = ds['Mon'].astype(int)
ds['Tue'] = ds['Tue'].astype(int)
ds['Wed'] = ds['Wed'].astype(int)
ds['Thu'] = ds['Thu'].astype(int)
ds['Sat'] = ds['Sat'].astype(int)
ds['Sun'] = ds['Sun'].astype(int)
#season
ds['Light Snow'] = ds['Light Snow'].astype(int)
ds['Mist + Cloudy'] = ds['Mist + Cloudy'].astype(int)
ds['spring'] = ds['spring'].astype(int)
ds['summer'] = ds['summer'].astype(int)
ds['winter'] = ds['winter'].astype(int)
ds['Year'] = ds['Year'].astype(int)
```

```python
ds.head()
```

```python
# Drop 'season','month','weekday','weathersit' as we have created the dummies for it
ds.drop(['season','Month','weekday','weathersit'], axis = 1, inplace = True)
ds.head()
```

## 4. Splitting the Data into training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
#np.random.seed(0)
ds_train, ds_test = train_test_split(ds, train_size = 0.7, test_size=0.3, random_state = 100)
```

```python
#rows and columns after split
print(ds_train.shape)
print(ds_test.shape)
```

# Rescaling the Features

```python
#Normalisation = (x-xmin)/(x max-x min)
#Standardisation= (x-mu)/ sigma
#import the library
from sklearn.preprocessing import MinMaxScaler
```

```python
#Instantiate an object
scaler = MinMaxScaler()

#Create a list of numeric variables
num_vars=['temp','Humidity','windspeed','count']

#Fit on data
ds_train[num_vars] = scaler.fit_transform(ds_train[num_vars])
ds_train.head()
```

```python
ds_train.describe()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated after scaling
#Little to no multicollinearity among predictors

plt.figure(figsize=(25, 20))
sns.heatmap(ds_train.corr(),cmap='YlOrRd',annot = True)
plt.show()
```

# Dividing X and Y sets for model building

```python
#Divide the data into X and y
y_train = ds_train.pop('count')
X_train = ds_train
```

## 5. Building a Linear Model

```python
#importing libraries
from sklearn.linear_model import  LinearRegression
from sklearn.feature_selection import RFE
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select=15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
#list of variable selected
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#Columns where RFE support is True
col = X_train.columns[rfe.support_]
col
```

```python
#Columns where RFE support is False
X_train.columns[~rfe.support_]
```

**Building model using statsmodel, for the detailed statistics** 

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model 
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

**Checking VIF**

```python
#drop the constant term B0
X_train_rfe = X_train_rfe.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
